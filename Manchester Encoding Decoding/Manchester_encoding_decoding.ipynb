```python
import matplotlib.pyplot as plt
```

```python
seq = "1001001010111"
encoded = []
```

```python
for c in seq:
    if c == '0':
        encoded += [-1,1]
    else :
        encoded += [1,-1]
```

```python
print(encoded)
```

```python
from turtle import width


plt.Figure(figsize=(12,4))
plt.step(range(len(encoded)),encoded,where="post")
plt.grid(True)
plt.title("Manchester Encoding")
plt.xticks(range(len(encoded)+1))
plt.yticks([-1,0,1],['Low(-1)','0','High(1)'])
plt.axhline(0, color='gray', linewidth=0.8)
plt.show()
```
