```python
import matplotlib.pyplot as plt 
```

```python
track = True
seq = "10110010"
encoded = []
inp = []
```

```python
seq = str(input())
```

```python
for c in seq:
    if c == '1':
        if track:
            track = False
            encoded.append(1)
        else :
            encoded.append(-1)
            track = True
        inp.append(1)
    else :
        encoded.append(0)
        inp.append(0)
```

```python
print(encoded)
```

```python
plt.Figure(figsize=(12,4))
plt.step(range(len(encoded)),inp,where="post")
plt.title("AMI Encoding Input")
plt.xlabel("Binary Bit")
plt.ylabel("Voltage Lavel")
plt.xticks(range(len(inp)+1))
plt.yticks([-1,0,1],["Low(-1)","0","High(+1)"])
plt.grid(True,linestyle = '--')
plt.show()





plt.Figure(figsize=(12,4))
plt.step(range(len(encoded)),encoded,where="post")
plt.title("AMI Encoding Output")
plt.xlabel("Binary Bit")
plt.ylabel("Voltage Lavel")
plt.xticks(range(len(encoded)+1))
plt.yticks([-1,0,1],["Low(-1)","0","High(+1)"])
plt.grid(True,linestyle = '--')
plt.show()
```

```python
output = []
```

```python
for i in encoded:
    if i == 1 or i == -1:
        output.append(1)
    else :
        output.append(0)
```

```python
print(output)
```

```python
plt.Figure(figsize=(12,4))
plt.step(range(len(encoded)),inp,where="post")
plt.title("AMI Decoding Input")
plt.xlabel("Binary Bit")
plt.ylabel("Voltage Lavel")
plt.xticks(range(len(inp)+1))
plt.yticks([-1,0,1],["Low(-1)","0","High(+1)"])
plt.grid(True,linestyle = '--')
plt.show()





plt.Figure(figsize=(12,4))
plt.step(range(len(output)),output,where="post")
plt.title("AMI Decoding Output")
plt.xlabel("Binary Bit")
plt.ylabel("Voltage Lavel")
plt.xticks(range(len(output)+1))
plt.yticks([-1,0,1],["Low(-1)","0","High(+1)"])
plt.grid(True,linestyle = '--')
plt.show()
```
