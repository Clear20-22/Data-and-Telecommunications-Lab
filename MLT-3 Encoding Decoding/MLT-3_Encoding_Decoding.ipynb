```python
import matplotlib.pyplot as plt 
```

```python

inpute = []
output = []
track = 1
seqoutput = [0,1,0,-1]
```

```python
seq = str(input())
```

```python
for c in seq:
    if c == '1':
        inpute.append(1)
        output.append(seqoutput[track%4])
        track += 1
    else:
        inpute.append(0)
        if track == 1:
            output.append(0)
        else:
            output.append(output[-1])

        
        
```

```python
print(output)
```

```python
plt.Figure(figsize=(12,4))
plt.step(range(len(output)),inpute,where="post")
plt.title("MLT-3 Encoding Input")
plt.xlabel("Binary Bit")
plt.ylabel("Voltage Lavel")
plt.xticks(range(len(inpute)+1))
plt.yticks([-1,0,1],["Low(-1)","0","High(+1)"])
plt.grid(True,linestyle = '--')
plt.show()





plt.Figure(figsize=(12,4))
plt.step(range(len(output)),output,where="post")
plt.title("MLT-3 Encoding Output")
plt.xlabel("Binary Bit")
plt.ylabel("Voltage Lavel")
plt.xticks(range(len(output)+1))
plt.yticks([-1,0,1],["Low(-1)","0","High(+1)"])
plt.grid(True,linestyle = '--')
plt.show()
```

```python
decode = []
check = [0]
for i in output:
    check.append(i)
```

```python
for i in range(1,len(check)):
    if check[i-1] == check[i]:
        decode.append(0)
    else:
        decode.append(1)
```

```python
print(decode)
```

```python
plt.Figure(figsize=(12,4))
plt.step(range(len(output)),inpute,where="post")
plt.title("MLT-3 Decoding Input")
plt.xlabel("Binary Bit")
plt.ylabel("Voltage Lavel")
plt.xticks(range(len(output)+1))
plt.yticks([-1,0,1],["Low(-1)","0","High(+1)"])
plt.grid(True,linestyle = '--')
plt.show()





plt.Figure(figsize=(12,4))
plt.step(range(len(decode)),decode,where="post")
plt.title("MLT-3 Decoding Output")
plt.xlabel("Binary Bit")
plt.ylabel("Voltage Lavel")
plt.xticks(range(len(decode)+1))
plt.yticks([-1,0,1],["Low(-1)","0","High(+1)"])
plt.grid(True,linestyle = '--')
plt.show()
```
