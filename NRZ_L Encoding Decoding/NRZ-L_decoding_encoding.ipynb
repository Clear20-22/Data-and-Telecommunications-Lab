```python
import matplotlib.pyplot as plt
```

```python
seq = "10110101"
encoded = []
```

```python
for c in seq:
    if c == '1':
        encoded.append(1)
    else :
        encoded.append(-1)
    
```

```python
print(encoded)
```

```python
plt.Figure(figsize=(10,4))
plt.step(range(len(encoded)),encoded,where="post")
plt.title("NRZ-L encoding")
plt.grid(True, linestyle='--', alpha=0.6)
plt.xlabel("Bit Index")
plt.ylabel("Voltage Lavel")
plt.axhline(0, color='gray', linewidth=0.8)
plt.xticks(range(len(seq)+1))
plt.yticks([-1,0,1],['Low(-1)','0','High(+1)'])
plt.show()
```
